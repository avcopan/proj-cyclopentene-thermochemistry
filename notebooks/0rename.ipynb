```python
from pathlib import Path

import automech

rmg_path = Path("../data/rmg")
```

```python
mech = automech.io.rmg.read.mechanism(
    rxn_inp=rmg_path / "cyclopentene_chem.inp",
    spc_inp=rmg_path / "cyclopentene_species_dictionary.txt",
)

print(automech.species_count(mech))
print(automech.reaction_count(mech))
```

```python
full_mech = automech.io.rmg.read.mechanism(
    rxn_inp=rmg_path / "full_cyclopentane_chem.inp",
    spc_inp=rmg_path / "full_cyclopentane_species_dictionary.txt",
)
```

```python
name_dct, *_ = automech.rename_dict(mech, full_mech)
mech = automech.rename(mech, names=name_dct)
```

```python
mech.species.write_json("0rename.json")
```
